## Data Visualization

```python
#Correlation Analysis between features
plt.figure(figsize=(10,10))
sub=np.triu(df_21.drop(["Country name","Regional indicator"],axis=1).corr())
sns.heatmap(df_21.drop(["Country name","Regional indicator"],axis=1).corr(), annot = True, fmt = ".2f", linewidth = .7,mask=sub)
plt.title("Relationship Between Features ")
plt.show()
```

```python
# Histogram of happiness scores in 2021
plt.figure(figsize=(10,10))
plt.hist(df_21['Ladder score'], bins=20)
plt.xlabel('Ladder score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Scores in 2021')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data = df_21, x = "Regional indicator")
plt.xticks(rotation = 60) 
plt.show()
```

```python
#distribution of feature set1
plt.figure(figsize=(10,10))
list_features1 = ["Social support", "Freedom to make life choices", "Generosity", "Perceptions of corruption"]
sns.boxplot(data = df_21.loc[:, list_features1], orient = "v", palette = "Set1") 
plt.show()
```

```python
# distribution of features set 2
plt.figure(figsize=(10,10))
list_features2 = ["Ladder score", "Logged GDP per capita"]
sns.boxplot(data = df_21.loc[:, list_features2], orient = "v", palette = "Set1") 
plt.show()
```

```python
# distribution of features set 3
plt.figure(figsize=(10,10))
list_features = ["Healthy life expectancy"]
sns.boxplot(data = df_21.loc[:, list_features], orient = "v", palette = "Set1")
plt.show()
```

```python
# Top 10 happiest countries
plt.figure(figsize=(10,10))
happiest=df_21[:10]
sns.barplot(x = "Ladder score", y = "Country name", data = happiest, palette = "coolwarm")
plt.title(" Top 10 Happiest Countries")
plt.show()
```

```python
# Top 10 unhappiest countries
plt.figure(figsize=(10,10))
happiest=df_21[-11:]
sns.barplot(x = "Ladder score", y = "Country name", data = happiest, palette = "coolwarm")
plt.title(" Top 10 Happiest Countries")
plt.show()
```

```python
import plotly.offline as pyo
pyo.init_notebook_mode(connected=True)
```

```python
#Ladder Score Distribution by country map
fig = px.choropleth(df.sort_values("year"), locations = "Country name", color =  "Life Ladder", locationmode = "country names", animation_frame = "year")
fig.update_layout(title = "Life Ladder Comparision by Country")
fig.show()
```

```python
#Most Generous and Ungenerous countries in 2021
data_generous =  df_21[(df_21.loc[:, "Generosity"] > 0.4) | (df_21.loc[:, "Generosity"] < -0.2)]
sns.barplot(x = "Generosity", y = "Country name", data = data_generous, palette = "coolwarm")
plt.show()
```

```python
#Relationship between Happiness by regional Indicator
plt.figure(figsize = (10, 10)) 
sns.kdeplot(data = df_21, x ="Ladder score", hue = "Regional indicator", fill = True, linewidth=2)
plt.axvline(df_21["Ladder score"].mean(), c = "black", linewidth = 3) 
plt.show()
```

```python
#Relationship between Corruption by regional score indicator
plt.figure(figsize = (10, 10)) 
sns.kdeplot(data = df_21, x ="Explained by: Generosity", hue = "Regional indicator", fill = True, linewidth=2)
plt.axvline(df_21["Explained by: Generosity"].mean(), c = "black", linewidth = 3) 
plt.show()
```

```python
plt.figure(figsize = (10, 10)) 
sns.kdeplot(data = df_21, x ="Explained by: Perceptions of corruption", hue = "Regional indicator", fill = True, linewidth=2)
plt.axvline(df_21["Explained by: Perceptions of corruption"].mean(), c = "black", linewidth = 3) 
plt.show()
```

```python
plt.figure(figsize = (20, 20)) 
sns.barplot(data = df_21, x = "Explained by: Freedom to make life choices", y = "Ladder score")
plt.xticks(rotation = 60)
plt.title(" Freedom vs Happiness")
plt.show()
```

```python
plt.figure(figsize = (20, 20)) 
sns.barplot(data = df_21, x = "Explained by: Log GDP per capita", y = "Ladder score")
plt.xticks(rotation = 60)
plt.title("Income vs Happiness")
plt.show()
```
